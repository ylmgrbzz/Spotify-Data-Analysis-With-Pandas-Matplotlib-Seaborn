```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("tracks.csv")
df.head()
```

```python
pd.isnull(df).sum()
```

```python
df.info()
```

```python
sorted_df=df.sort_values('popularity', ascending=True).head(10)
sorted_df
```

```python
df.describe().transpose()
```

```python
most_popular=df.query('popularity > 90',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
df.set_index("release_date", inplace=True)
df.index=pd.to_datetime(df.index)
df.head()
```

```python
df[["artists"]].iloc[18]
```

```python
df["duration"]=df["duration_ms"].apply(lambda x: round(x/100))
df.drop("duration_ms",inplace=True, axis=1)
```

```python
df.duration.head()
```

```python
corr_df=df.drop(["key","mode","explicit"],axis=1).corr("pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=1 , vmax=1, center=0, linewidths=1 , cmap="inferno", linecolor="Black")
heatmap.set_title("Değişken Arasındaki Korelasyon Isı Haritası")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df.sample(int(0.004*len(df)))
```

```python
len(sample_df)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness", x="energy",color="c").set(title="Ses Yüksekliği Ve Enerji Korelasyonu")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity", x="acousticness",color="b").set(title="Popülerite Ve Akustik Korelasyonu")
```

```python
df["dates"]=df.index.get_level_values("release_date")
df.dates=pd.to_datetime(df.dates)
years=df.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Yıllık Şarkı Sayısı")
```

```python
total=df.duration
figg=(18,7)
fig,ax=plt.subplots(figsize=figg)
fig=sns.barplot(x=years,y=total, ax=ax , errwidth=False).set(title="Yıllar ve Süre")
plt.xticks(rotation=90)
```

```python
total=df.duration
sns.set_style(style="whitegrid")
figg=(10,5)
fig,ax=plt.subplots(figsize=figg)
fig=sns.lineplot(x=years, y=total , ax=ax).set(title="Yıllar ve Süre")
plt.xticks(rotation=60)
```
